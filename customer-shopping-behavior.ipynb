```python
df['age_group'] = pd.cut(df['age'],
    bins=[0,18,30,45,60,100],
labels=['Teen','Young Adult','Adult','Mid Age','Senior'])
```

```python
df[['age', 'age_group']].head(10)
```

```python
freq_map = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(freq_map)
```
